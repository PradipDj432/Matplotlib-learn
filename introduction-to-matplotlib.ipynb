# Matplotlib workflow

```python
# import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# Prepare data
x = [1,2,3,4,5]
y = [11,22,33,44,55]

# setup plot
fig,ax = plt.subplots(figsize=(10,10))#(width,height)

# plot data
ax.plot(x,y)

# Customise plot
ax.set(title="simple plot",\
      xlabel="x-axis",
      ylabel="y-axis")
# save & show
fig.savefig("numpy-images/sample-plot.png")
```

# Subplot

```python
# 1st methode
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,
                                        figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nbp.keys(),nbp.values());
ax4.hist(x,20);
```

```python
# 2nd methode
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nbp.keys(),nbp.values());
ax[1,1].hist(x,20);
```

## which one should you use ? (pyplot or matplotlib object orinted)
* Adavans plotting than oo use
* Quick and expriment than use pyplot

```python
over_50 = hd[hd["age"]>50]
```

```python
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,5));
```

```python
# oo method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,5))
over_50.plot(kind = "scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(10,5))
over_50.plot(kind = "scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,100]);
```

```python
# oo method from scratch 
fig , ax = plt.subplots(figsize=(10,6))
over_50_plot = ax.scatter(x=over_50['age'],
                       y=over_50['chol'],
                       c=over_50['target']);

# customize
ax.set(title = "heart disease",
       xlabel = "age",
       ylabel = "chol")

# add a legend
ax.legend(*over_50_plot.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--')
```

```python
over_50.target.values
```

```python
# oo method from scratch 
fig , ax = plt.subplots(figsize=(10,6))
over_50_plot = ax.scatter(x=over_50['age'],
                       y=over_50['chol'],
                       c=over_50['target']);

# customize
ax.set(title = "heart disease",
       xlabel = "age",
       ylabel = "chol")

# add a legend
ax.legend(*over_50_plot.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--')
```

```python
# subplot of chol ,age, htalach


fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,
                             figsize=(10,10),
                             sharex=True);

# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                           y=over_50["chol"],
                           c=over_50["target"]);

# customisze ax0

ax0.set(title="heart disesse age vs chol",
        ylabel="chol");

# add legend to ax0
ax0.legend(*scatter.legend_elements(),title="target");

# add mean line ax0
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                           y=over_50["thalach"],
                           c=over_50["target"]);

# customisze ax1

ax1.set(title="heart disesse age vs thalach",
        xlabel="age",
        ylabel="thalach");

# add legend to ax1
ax1.legend(*scatter.legend_elements(),title="target");

# add mean line ax1
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");

## add title to fig
fig.suptitle("Heart disease analysis",fontsize=16,fontweight="bold");
```
